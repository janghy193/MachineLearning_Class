```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('insurance.csv')
display(df)

enc = OneHotEncoder()
# 값 추출, 2차원 배열화
gender_arr = df['gender'].values.reshape(-1,1)
smoker_arr = df['smoker'].values.reshape(-1,1)
region_arr = df['region'].values.reshape(-1,1)

# 기존의 세 컬럼 제거
del df['gender']
del df['smoker']
del df['region']

# 숫자컬럼으로 변형, 데이터프레임화
# encoded = enc.fit_transform(df[['gender','smoker','region']]).toarray()
# col_names = enc.get_feature_names(['gender','smoker','region'])
# df2 = pd.DataFrame(encoded, columns=col_names)
gender_enc = enc.fit_transform(gender_arr).toarray()
g_name_list = enc.get_feature_names(['gender'])
gender_df = pd.DataFrame(gender_enc,columns=g_name_list)

smoker_enc = enc.fit_transform(smoker_arr).toarray()
s_name_list = enc.get_feature_names(['smoker'])
smoker_df = pd.DataFrame(smoker_enc,columns=s_name_list)

region_enc = enc.fit_transform(region_arr).toarray()
r_name_list = enc.get_feature_names(['region'])
region_df = pd.DataFrame(region_enc,columns=r_name_list)

# 합치기
df = pd.concat([df,gender_df,smoker_df,region_df], axis=1)
display(df)

# --------------- 2번 -----------------
# 학습용, 검증용으로 분리
x = df.drop('charges',axis=1)
y = df['charges']

# 학습용 (1100개)
x_train = x[:1100]
label_train = y[:1100]

# 검증용 (238개)
x_test = x[1100:]
label_test = y[1100:]

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, input_dim=11, activation = 'linear')) # input_dim= features 수(컬럼 수)
model.add(Dense(32, input_dim=11, activation = 'linear'))
model.add(Dense(16, input_dim=11, activation = 'linear'))
model.add(Dense(8, input_dim=11, activation = 'linear'))
model.add(Dense(4, input_dim=11, activation = 'linear'))
model.add(Dense(1, input_dim=11, activation = 'linear'))

model.compile(loss='mean_absolute_error', optimizer='adam',metrics=['accuracy'])

model.fit(x_train, label_train, batch_size=64, epochs=60,validation_data=(x_test, label_test))

model_summary = pd.DataFrame(np.array(model.evaluate(x_test,label_test)).reshape(1,2), columns = model.metrics_names)
display(model_summary)

# ----------- 3번 --------------
print('\n-----테스트데이터에 대한 예측 3회 실행-----')
for i in range(3):
    print(i+1,'회차 예측')
    # 테스트용 데이터 행 번호 랜덤 추출
    num = np.random.randint(1100,len(df),1)[0]
    
    # 테스트용 라벨 한 개 추출
    pick = label_test[num]
    print('실제:', pick)
    
    # 추출한 행 번호에 해당하는 데이터로 예측 수행
    pred = np.mean(model.predict(x_test.loc[[num]]).ravel())    # 1차원으로 축소 후 모든 결과의 평균값 산출
    print('추정:', pred)
    print('오차:',pick-pred,'\n')
```

```python
# 다중분류

import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

df = pd.read_csv('mc_train.csv')
display(df)

x_train = df.drop(['class'],axis=1).loc[:1099]
x_test = df.drop(['class'],axis=1).loc[1100:]
label_train = df['class'][:1100]
label_test = df['class'][1100:]

# print(len(df['class'].unique()))  # 6

# 딥러닝
act = 'softmax'
model = Sequential()
model.add(Dense(64, input_dim=11, activation = act)) # input_dim= features 수(컬럼 수)
model.add(Dense(32, activation = act))
model.add(Dense(10, activation = act))
#model.add(Dense(8, activation = act))
#model.add(Dense(4, activation = act))
#model.add(Dense(1, activation = act))

model.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop',metrics=['accuracy'])
#model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])

model.fit(x_train, label_train, batch_size=64, epochs=10,validation_data=(x_test, label_test), verbose=0)

model_summary = pd.DataFrame(np.array(model.evaluate(x_test,label_test)).reshape(1,2), columns = model.metrics_names)
display(model_summary)

print('\n-----테스트데이터에 대한 예측 3회 실행-----')
for i in range(3):
    print(i+1,'회차 예측')
    # 테스트용 데이터 행 번호 랜덤 추출
    num = np.random.choice(label_test.index,1)[0]
    
    # 테스트용 라벨 한 개 추출
    pick = label_test[num]
    print('실제:', pick)
    
    # 추출한 행 번호에 해당하는 데이터로 예측 수행
    pred = model.predict(x_test.loc[[num]]).ravel()    # 1차원으로 축소 후 모든 결과의 평균값 산출
    print('추정결과:', np.argmax(pred)+1)     # 가장 큰값이 몇 번째인지(0부터 세기) +1
    print('추정확률값:')
    display(pd.DataFrame(pred,columns=['prob'], index=np.arange(1,11)).T)

top_2 = df.corr().abs().nlargest(3,'class').drop('class')
print(top_2.index[0], top_2.index[1])
col1 = top_2.index[0]
col2 = top_2.index[1]

# 두 컬럼에 대한 전체 데이터 플롯
plt.scatter(df[col1], df[col2], c = df['class'], s=2.5)
plt.show()

# 테스트용 데이터에 대한 예측값 산출
pred = model.predict(np.c_[df[col1],df[col2]])
pred_res = []
for p in pred:
    pred_res.append(np.argmax(p)+1)

# 예측결과 플롯
plt.scatter(df[col1], df[col2], c = pred_res, s=2.5)

#plt.legend()
plt.xlabel(col1)
plt.ylabel(col2)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

df = pd.read_csv('Social_Network_Ads.csv')

# Age와 Purchased 간 로지스틱 회귀분석
x = df['Age'].to_numpy().reshape(-1,1)
y = df['Purchased']

plt.scatter(x,y,c=df['Purchased'])


from sklearn import linear_model
log_regress = linear_model.LogisticRegression()
log_regress.fit(X = x, y = y)
print('\n--로지스틱 모델링--')
print('로지스틱 절편:',log_regress.intercept_[0])
print('로지스틱 계수:',log_regress.coef_[0][0])

print('\n---값 예측----')
predx = 46
res = log_regress.predict([[predx]])[0]
prob = max(log_regress.predict_proba([[predx]])[0])
print(str(predx)+'세 구매 여부 예측 결과:',res)
print('확률:',round(prob,2)*100,'%')

# 시그모이드 곡선 정의
def sigmoid(x):
    return (1 / (1 + np.exp(-(log_regress.intercept_[0] +
                              (log_regress.coef_[0][0] * x)))))
# 시그모이드 곡선 데이터 정의
x1 = np.arange(min(x), max(x), 1)
y1 = [sigmoid(n) for n in x1]
# 값 플로팅
plt.scatter(x,y,facecolors='none', c=y)
# 예측값 플로팅
plt.plot(predx,prob,marker='*',color='r')
# 점선 그리기
predx_linex = np.arange(min(x),predx,1)
predx_liney = np.arange(0,prob,0.01)
plt.plot(predx_linex, predx_linex*0+prob, '--', color='gray', alpha = 0.25)
plt.plot(predx_liney*0+predx, predx_liney, '--', color='gray', alpha = 0.25)

# 로지스틱 회귀모델 (시그모이드 곡선) 플로팅
plt.plot(x1,y1,c='m')
# 점선(0.5 확률 구분선)
#plt.plot(x1,x1*0+0.5,'b--',alpha = 0.25)

yel = mpatches.Patch(color='y', label='Purchased')
pur = mpatches.Patch(color='purple', label='Not Purchased')
plt.legend(handles=[yel, pur],bbox_to_anchor=(1.4, 1) , loc='upper right')
plt.title('Age - Purchased Plot')
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

df = pd.read_csv('Social_Network_Ads.csv')
del df['User ID']
df['Gender'][df['Gender']=='Male'] = 0
df['Gender'][df['Gender']=='Female'] = 1
df['Gender'] = pd.to_numeric(df['Gender'])
df_feat = df.iloc[:,0:3]
from sklearn.model_selection import train_test_split
train_set, test_set, train_labels, test_labels = train_test_split(
    df_feat, # features
    df['Purchased'], # labels
    test_size = 0.5, # split ratio
    random_state = 1, # set random seed
    
    stratify = df['Purchased']) # randomize based on labels
x = train_set # mean radius
y = train_labels  # 0, 1
log_regress = linear_model.LogisticRegression()
log_regress.fit(X = x, y = y)
print('\n--전체 컬럼에 대한 예측--')
print('로지스틱 절편:',log_regress.intercept_)
print('로지스틱 계수:\n',log_regress.coef_)
testset = [[1,1,160000]]
print('모델 예측 테스트:',testset[0],'\n결과:',log_regress.predict_proba(testset),log_regress.predict(testset))
#---get the predicted probablities and convert into a dataframe---
preds_prob = pd.DataFrame(log_regress.predict_proba(X=test_set))
#---assign column names to prediction---
preds_prob.columns = ["Not purchased", "Purchased"]
#---get the predicted class labels---
preds = log_regress.predict(X=test_set)
preds_class = pd.DataFrame(preds)
preds_class.columns = ["Prediction"]
#---actual diagnosis---
original_result = pd.DataFrame(test_labels)
original_result.columns = ["Original Result"]
original_result.reset_index(drop=True,inplace=True)
#---merge the three dataframes into one---
result = pd.concat([preds_prob, preds_class, original_result], axis=1)
print('\n산출 예측 결과 상위5개:')
print(result.head())

#---generate table of predictions vs actual---
print("---Confusion Matrix---")
print(pd.crosstab(preds, test_labels))

#---get the accuracy of the prediction---
print("---Accuracy---")
print(log_regress.score(X = test_set, y = test_labels))

# View summary of common classification metrics
from sklearn import metrics

print("---Metricses---")
print(metrics.classification_report(y_true = test_labels, y_pred = preds))
```

```python
# ROC곡선
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

df = pd.read_csv('Social_Network_Ads.csv')

from sklearn.model_selection import train_test_split
train_set, test_set, train_labels, test_labels = train_test_split(
    df['Age'].to_numpy().reshape(-1,1), # features
    df['Purchased'], # labels
    test_size = 0.5, # split ratio
    random_state = 1, # set random seed
    
    stratify = df['Purchased']) # randomize based on labels
x = train_set # mean radius
y = train_labels  # 0, 1
log_regress = linear_model.LogisticRegression()
log_regress.fit(X = x, y = y)

from sklearn.metrics import roc_curve, auc
#---find the predicted probabilities using the test set
probs = log_regress.predict_proba(test_set)
preds = probs[:,1]
#---find the FPR, TPR, and threshold---
fpr, tpr, threshold = roc_curve(test_labels, preds)

print('FPR:\n', fpr)
print('TPR:\n', tpr)
print('Threshhold:\n', threshold)

#---find the area under the curve---
roc_auc = auc(fpr, tpr)
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate (TPR)')
plt.xlabel('False Positive Rate (FPR)')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc = 'lower right')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns; sns.set(font_scale=1.2)
import matplotlib.pyplot as plt
data = pd.read_csv('svm.csv')
sns.lmplot('x1', 'x2',
 data=data,
 hue='r',
 palette='Set1',
 fit_reg=False,
 scatter_kws={"s": 50});
from sklearn import svm
#---Converting the Columns as Matrices---
points = data[['x1','x2']].values
result = data['r']
clf = svm.SVC(kernel = 'linear')
clf.fit(points, result)
print('Vector of weights (w) = ',clf.coef_[0])
print('b = ',clf.intercept_[0])
print('Indices of support vectors = ', clf.support_)
print('Support vectors = ', clf.support_vectors_)
print('Number of support vectors for each class = ', clf.n_support_)
print('Coefficients of the support vector in the decision function = ',
 np.abs(clf.dual_coef_))
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles
#---X is features and c is the class labels---
X, c = make_circles(n_samples=500, noise=0.09)
rgb = np.array(['r', 'g'])
plt.scatter(X[:, 0], X[:, 1], color=rgb[c])
plt.show()
fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')
z = X[:,0]**2 + X[:,1]**2
ax.scatter(X[:, 0], X[:, 1], z, color=rgb[c])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# 강사님이 소개해 주신 알고리즘 - 0~1의 소수 리스트를 총합이 1이 되도록 가공해주는 알고리즘
def get_ratio_alg(num):
    default = [0,100]
    rand = np.random.randint(0,100,num).tolist()
    
    for i in rand:
        default.append(i)
    
    default.sort()
    
    for i in range(len(default)-1,-1,-1):
        if i != 0:
            default[i] = default[i] - default[i-1]
    
    del default[0]
    
    return default


res2 = get_ratio_alg(10)
print(res2, sum(res2))
plt.pie(res2, shadow=True, explode = [0,0.2,0,0,0,0,0,0,0,0,0], startangle=45, autopct="%.lf")
plt.title("piechart")
plt.show()
```
